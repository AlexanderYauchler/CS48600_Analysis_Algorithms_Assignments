##Load the graph from a file to adjacency list

```python
file_path = './graph.txt'

adjacency_list = {}
with open(file_path, 'r') as file:
    for line in file:
        vertices = line.strip().split(' ')
        index = int(vertices[0]) - 1
        adjacency_list[index] = []
        for i in range(1, len(vertices)):
                adjacency_list[index].append(int(vertices[i])-1)

source = int('1') - 1

print("Adjacency List: ")
print(adjacency_list)
print("source: " + str(source))
```

##BFS and Shortest Path
(Refer to Section 8.2)

```python
import math
from collections import deque

def augmented_BFS(adjacency_list, source):
    adjCount = len(adjacency_list)

    dist = [math.inf] * adjCount

    parent = [None] * adjCount

    queue = deque()
    dist[source] = 0
    queue.append(source)

    while queue:
        u = queue.popleft()
        for v in adjacency_list.get(u, []):
            if dist[v] is math.inf:
                dist[v] = dist[u] + 1
                parent[v] = u
                queue.append(v)

    return [(dist[i], parent[i]) for i in range(adjCount)]

BFS_result = augmented_BFS(adjacency_list, source)
for i in range(len(BFS_result)):
    print(str(i+1) + ": " + str(BFS_result[i]))
```

```python
# Implement Knapsack Algorithm and Knapsack Reconstruction
# Be careful with the index of arrays

def knapsack(values, sizes, capacity):
    count = len(values)
    table = [[0] * (capacity + 1) for _ in range(count + 1)]

    for item in range(1, count + 1):
        for room in range(capacity + 1):
            if sizes[item - 1] > room:
                table[item][room] = table[item - 1][room]
            else:
                table[item][room] = max(table[item - 1][room],
                                        table[item - 1][room - sizes[item - 1]] + values[item - 1])

    return table

def knapsack_reconstruct(values, sizes, capacity, A):
    count = len(values)
    chosen = []
    room = capacity

    for item in range(count, 0, -1):
        if sizes[item - 1] <= room and \
           A[item - 1][room - sizes[item - 1]] + values[item - 1] >= A[item - 1][room]:
            chosen.append(item - 1)
            room -= sizes[item - 1]

    chosen.reverse()
    return chosen
```

```python
# Load test data from the file

def load_data(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        first_line = lines[0].strip()
        C = int(first_line.split()[0])
        n = int(first_line.split()[1])
        values = []
        sizes = []

        for i in range(1, n + 1):
            value, size = map(int, lines[i].strip().split())
            values.append(value)
            sizes.append(size)

    return C, values, sizes
```

```python
# run on the testing case

filename = 'testcase.txt'
C, values, sizes = load_data(filename)
print("Knapsack capacity: ", C)
print("Number of items: ", len(values))
print("Values: ", values)
print("Sizes: ", sizes)
```

```python
A = knapsack(values, sizes, C)
print("Solution to the Knapsack problem: ", A[len(values)][C])

S = knapsack_reconstruct(values, sizes, C, A)
print("Items added to the knapsack: ", S)
print("Total size of the knapsack: ", sum([sizes[i] for i in S]))
print("Total value of the knapsack: ", sum([values[i] for i in S]))
```
